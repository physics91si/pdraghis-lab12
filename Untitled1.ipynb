```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
def generate_data(N=30000, num_missing=700):
    x = np.random.uniform(0,75,N)
    y = np.random.uniform(0,75,N)
    z = np.random.uniform(0,75,N)
    mstar = 10 **np.random.uniform(8,12,N)
    shuffled_idxs = np.random.permutation(N)
    mstar[shuffled_idxs[:num_missing]] = 0
    sfr = np.random.uniform(0,1e-9,N)
    sfr[shuffled_idxs[num_missing:7*num_missing]] = 0
    mvir = 10 ** np.random.uniform(9,13,N)
    return pd.DataFrame({'x': x,
                         'y': y,
                         'z': z,
                         'sfr': sfr,
                         'mstar': mstar,
                         'mvir': mvir})
df = generate_data()
```

```python
central_slice = df[df.z.between(33,37)]
plt.hexbin(central_slice.x, central_slice.y, bins='log', cmap='coolwarm')
plt.colorbar()
```

```python
#impose the physical constraint of having stellar mass inside a galaxy
df=df[df.mstar>0]

#specific star formation rate == SFR/mstar
df['ssfr']=np.log10(df.sfr/df.mstar)
```

```python
df['ssfr'] = df.apply(lambda x: np.log10(x['sfr']/x['mstar']) if x['sfr']>0 else -12, axis=1)
```

```python
df.to_csv('galaxies.csv')
```
